```python
df = data.drop(["ï»¿UDI", "Product ID" , "TWF" , "TWF" , "PWF" , "OSF" ,"RNF","HDF"] , axis = 1)
```

```python
df.head()
```

```python
X = df.drop('Machine failure', axis=1)
y = df['Machine failure']
```

```python
cat_f = ['Type']
num_f = [
    'Air temperature [K]',
    'Process temperature [K]',
    'Rotational speed [rpm]',
    'Torque [Nm]',
    'Tool wear [min]'
]
```

```python
X_train ,X_test , y_train , y_test =train_test_split(X,y , random_state=42,stratify=y)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(drop='first'), cat_f),
        ('num', 'passthrough', num_f)
    ]
)
```

```python
x_transformed = preprocessor.fit_transform(X_train)
x_test_trans = preprocessor.transform(X_test)
```

```python
from imblearn.over_sampling import SMOTE

abc = SMOTE(random_state=42)
X_res, y_res = abc.fit_resample(x_transformed, y_train)
```

```python
from sklearn.ensemble import BaggingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression(max_iter=1000)
lr_l1 = LogisticRegression(l1_ratio=1, solver= "saga",max_iter=1000,penalty="elasticnet")
lr_l2 = LogisticRegression(l1_ratio=0,solver="saga",max_iter=1000,penalty="elasticnet")
```

```python
Log_pipeline = Pipeline([("scaler" , scaling),("model",log)])
lr_l1_pip = Pipeline([("scaler" , scaling),("model",lr_l1)])
lr_l2_pip = Pipeline([("scaler" , scaling),("model",lr_l2)])
xgb = XGBClassifier(
    learning_rate=0.03,
    n_estimators=700,
    max_depth=5,
    random_state=42
)

cat = CatBoostClassifier(
    learning_rate=0.03,
    iterations=700,
    depth=5,
    random_state=42,
    verbose=0
)
```

```python
models = {"Forest" : RandomForestClassifier(),
         "SVM" : svm_pipeline,
         "Tree" : DecisionTreeClassifier(),
         "logistic" : log,
         "L!":lr_l1_pip,
         "l2":lr_l2_pip,
         "Bagging": BaggingClassifier(),
         "XG" : xgb,
         "CAT" :cat,
}
```

```python
res=[]
```

```python
res
```

```python
res.clear()
```

```python
for name , model in models.items():
    model.fit(X_res,y_res)
    y_pred = model.predict(x_test_trans)

    cm =confusion_matrix(y_test,y_pred)
    aas=accuracy_score(y_test,y_pred)
    ps=precision_score(y_test,y_pred)
    rs=recall_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)
    res.append({"Model": name,
                   "Confusion Matrix" : cm,
                   "Accuracy":aas,
                   "Precision ": ps,
                   "recall": rs,
                   "F1 score" : f1
                  })

    
```

```python
df = pd.DataFrame(res)
```
